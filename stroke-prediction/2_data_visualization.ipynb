# ***Data Visualization***

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing Dataset

```python
df = pd.read_csv("data/cleaned-stroke-dataset.csv")
df.head()
```

---

## Correlation Heatmap

```python
import seaborn as sn
plt.figure(figsize = (16, 8))
sn.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

- The graph shows that age has the highest corr with stroke. As the age increases, chances of getting a stroke increases.
- After age, the factors that affects stroke the most are hypertension, heart disease, and glucose level. Thus it can be concluded that people with tension problems, glucose problem and heart diseases have chances of getting a stroke.

---

## Pie Chart of Stroke Counts

```python
plt.figure(figsize=(12, 6))
plt.pie(df["stroke"].value_counts(), 
        labels=["Not Stroke", "Stroke"], autopct="%.2f", colors=["orange", "aqua"])
plt.title("Pie Chart of Stroke")
plt.show()
```

This clearly shows that the dataset we are using is highly imbalanced. To tackle this, we will use SMOTE.

---

## Bar Chart : Smoking Status vs Stroke

```python
stroke_by_smoking = df.groupby("smoking_status")["stroke"].mean() * 100
print(stroke_by_smoking)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(stroke_by_smoking.index, stroke_by_smoking.values)
plt.xlabel("Smoking Status")
plt.ylabel("Stroke Ratio")
plt.title("Stroke vs Smoking Status")
plt.show()
```

The graph shows that the people who smoked in the past or still smoke have higher chances of getting a stroke than from those who don't smoke.

---

## Stroke by Age

```python
stroke_by_age = df.groupby("age")["stroke"].mean() * 100

plt.figure(figsize = (12, 6))
plt.plot(stroke_by_age.index, stroke_by_age.values, color = "red",marker = "o")
plt.xlabel("Age")
plt.ylabel("Stroke Ratio")
plt.title("Stroke Risk by Age", fontsize=15)
plt.show()
```

- The graph shows that the risk of getting a stroke is higher in newborns and in old people. The increase in the risk is high after the age of 50.
- Teenagers are mostly unaffected by stroke.

---

## Stroke by Heart Disease

```python
stroke_by_heart_disease = df.groupby("heart_disease")["stroke"].mean() * 100
stroke_by_heart_disease
```

```python
plt.figure(figsize = (12, 6))
plt.bar(stroke_by_heart_disease.index, stroke_by_heart_disease.values)
plt.xlabel("Patient's Heart Disease Status")
plt.ylabel("Stroke Ratio")
plt.title("Stroke Risk by Heart Disease", fontsize=15)
plt.xticks([0, 1], ["No Heart Problem", "Heart Problem"])
plt.show()
```

The graph shows people with heart disease have higher chances of getting a stroke than people with no heart disease.
