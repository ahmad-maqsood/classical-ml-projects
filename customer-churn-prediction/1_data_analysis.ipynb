The dataset is available on Kaggle : https://www.kaggle.com/datasets/blastchar/telco-customer-churn .

## Data Preprocessing

```python
df["Partner"] = df["Partner"].map({"Yes" : 1, "No" : 0})
df["Dependents"] = df["Dependents"].map({"Yes" : 1, "No" : 0})
df["PhoneService"] = df["PhoneService"].map({"Yes" : 1, "No" : 0})
df["MultipleLines"] = df["MultipleLines"].map({"Yes" : 1, "No" : 0, "No phone service" : 0})
df["OnlineSecurity"] = df["OnlineSecurity"].map({"Yes" : 1, "No" : 0, "No internet service" : 0})
df["OnlineBackup"] = df["OnlineBackup"].map({"Yes" : 1, "No" : 0, "No internet service" : 0})
```

```python
print(f'Unique Values in the "Partner" col. : {(df["Partner"].unique()).tolist()}')
print(f'Unique Values in the "Dependents" col. : {(df["Dependents"].unique()).tolist()}')
print(f'Unique Values in the "PhoneService" col. : {(df["PhoneService"].unique()).tolist()}')
print(f'Unique Values in the "MultipleLines" col. : {(df["MultipleLines"].unique()).tolist()}')
print(f'Unique Values in the "OnlineSecurity" col. : {(df["OnlineSecurity"].unique()).tolist()}')
print(f'Unique Values in the "OnlineBackup" col. : {(df["OnlineBackup"].unique()).tolist()}')
```

```python
df["DeviceProtection"] = df["DeviceProtection"].map({"Yes" : 1, "No" : 0, "No internet service" : 0})
df["TechSupport"] = df["TechSupport"].map({"Yes" : 1, "No" : 0, "No internet service" : 0})
df["StreamingTV"] = df["StreamingTV"].map({"Yes" : 1, "No" : 0, "No internet service" : 0})
df["StreamingMovies"] = df["StreamingMovies"].map({"Yes" : 1, "No" : 0, "No internet service" : 0})

df["PaperlessBilling"] = df["PaperlessBilling"].map({"Yes" : 1, "No" : 0})
df["Churn"] = df["Churn"].map({"Yes" : 1, "No" : 0})
```

```python
print(f'Unique Values in the "DeviceProtection" col. : {(df["DeviceProtection"].unique()).tolist()}')
print(f'Unique Values in the "TechSupport" col. : {(df["TechSupport"].unique()).tolist()}')
print(f'Unique Values in the "StreamingTV" col. : {(df["StreamingTV"].unique()).tolist()}')
print(f'Unique Values in the "StreamingMovies" col. : {(df["StreamingMovies"].unique()).tolist()}')
print(f'Unique Values in the "PaperlessBilling" col. : {(df["PaperlessBilling"].unique()).tolist()}')
print(f'Unique Values in the "Churn" col. : {(df["Churn"].unique()).tolist()}')
```

Okay...so...there are some blank spaces. Maybe between the amounts? let's see

```python
df[df["TotalCharges"].str.contains(" ")]
```

If you look at it, all these total charges have the contract of 2 years. We can just multiply the monthly payment by 24 for these.

But first, we need to convert these spaces to NaN

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce") #errors="coerce" turns spaces into NaN
```

```python
mask = df["TotalCharges"].isna()
df.loc[mask, "TotalCharges"] = df.loc[mask, "MonthlyCharges"] * 24
```
